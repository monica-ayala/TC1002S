```python
import seaborn as sb
import numpy as np; np.random.seed(0)
from matplotlib import cm
```

```python
df = pd.read_csv('../datasets/comprar_alquilar.csv')
sb.set_theme(style="whitegrid")
ax = sb.boxplot(x=df['gastos_comunes'])
```

```python
sb.set_theme(style="whitegrid")
iris= sb.load_dataset("iris")
ax = sb.boxplot(x=iris["sepal_length"])
```

```python
plt.figure(figsize=(12,12))
ax = sb.boxplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
plt.figure(figsize=(12,12))
ax = sb.boxplot(x="sepal_length", y="sepal_width", hue="species",
                 data=iris, palette="Set2")
```

```python
plt.figure(figsize=(12,12))
g = sb.catplot(x="sepal_length", y="petal_length",
                hue="species", col="species",
                data=iris, kind="box",
                height=5, aspect=.7);
```

```python
colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sb.heatmap(iris.corr(), linewidths=0.1, vmax=1.0, square=True, 
           cmap=colormap, linecolor='white', annot=True);
```

```python
normal_data = np.random.randn(20, 35)
ax = sb.heatmap(normal_data, center=2)
```
