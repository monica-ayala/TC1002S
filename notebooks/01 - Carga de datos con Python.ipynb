```python
import csv
with open ('../datasets/msft.csv', encoding='utf-8') as csvFile:
    sample = csvFile.read(64)
    has_header = csv.Sniffer().has_header(sample)
    print(has_header)

    deduced_dialect = csv.Sniffer().sniff(sample)
```

```python
with open ('../datasets/msft.csv', encoding='utf-8') as csvFile:
    reader = csv.reader(csvFile, deduced_dialect)
    for row in reader:
        print(row)
```
